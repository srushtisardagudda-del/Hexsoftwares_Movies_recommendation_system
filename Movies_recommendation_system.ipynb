```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/dataset (1).csv')
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags'] = movies['genre'] + movies['overview']
movies.head()
```

```python
new_df = movies[['id', 'title', 'tags','genre','overview']]
new_df.head()
```

```python
new_df = new_df.drop(columns = ['genre','overview'])
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Shawshank Redemption']
```

```python
dist = sorted(list(enumerate(sim[2])), reverse=True, key=lambda vector:vector[1])
dist
```

```python
for i in dist[0:5]:
  print(new_df.iloc[i[0]].title)
```

```python
from scipy.spatial import distance
def recommend(movies):
  index = new_df[new_df['title']==movies].index[0]
  distancest = sorted(list(enumerate(sim[index])), reverse=True, key=lambda vec:vec[1])
  for i in distancest[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
